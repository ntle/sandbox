## Solving the problem

### Loading data

```python
# pd.DataFrame(data1 + data2).to_csv("/home/le/Dropbox/nintankle_ntle/data.csv", header=False, index=False)

dropbox_dir = "/home/le/Dropbox/nintankle_ntle/SDL/"
data_src = dropbox_dir  +'HW3/data.csv'

data_df = pd.read_csv(data_src, names=['x', 'y'])
data_df.head()
```

### K-Means Solution

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(data_df.values) 
```

```python
#means
kmeans.cluster_centers_ 
```

```python
#assign labels to each datapoint
data_df['label'] = kmeans.labels_
data_df.head()
```

#### Drawing results

```python
cluster1_elements = data_df[data_df['label'] == 0]
cluster2_elements = data_df[data_df['label'] == 1]

plt.scatter(cluster1_elements['x'], cluster1_elements['y'], c='b', marker='+')
plt.scatter(cluster2_elements['x'], cluster2_elements['y'], c='r', marker='+')
```

### GMM Solution

```python
from sklearn.mixture import GMM

gmm = GMM(n_components=2, covariance_type='full')
gmm.fit(data_df.values) 
```

```python
#centers
gmm.means_ 
```

```python
#convariance matrixes
gmm.covars_
```

#### Drawing results

```python
from matplotlib.patches import Ellipse

#plot data
ax = data_df.plot(x='x', y='y', kind='scatter', marker='.')

#plot ellipses
for i in range(len(gmm.means_)):
    mean = gmm.means_[i]
    covariance = gmm.covars_[i]
    
    ax.add_artist(Ellipse(xy=mean, width=covariance[0][0], height=covariance[1][1], facecolor=rand(3)))


```
